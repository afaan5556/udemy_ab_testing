### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

### Constants

```python
NUM_TRIALS = 2000
BANDIT_PROBABILITIES = [0.2, 0.5, 0.75]
# BANDIT_PROBABILITIES = [0.2, 0.71, 0.76]
```

### Bandit Class

```python
class Bandit():
    def __init__(self, p):
        self.p = p
        self.a = 1 # Start with uniform disstribution given no prior information
        self.b = 1 # Start with uniform disstribution given no prior information
        self.N = 0 # Info only
    
    def pull(self):
        return np.random.random() < self.p

    def sample(self):
        return np.random.beta(self.a, self.b) # Use numpy random to get a sample of current beta
        # Note, using the beta module directly here would have been ok too

    def update(self, x):
        self.a += x
        self.b += 1 - x 
        self.N += 1
```

### Plot Function

```python
def plot(bandits, trial):
    x = np.linspace(0, 1, 200)
    for b in bandits:
        y = beta.pdf(x, b.a, b.b)
        try:
            wr = (b.a - 1)/b.N
        except:
            wr = 'not played yet'
        plt.plot(x, y, label="Real P: {}:.4f, Win rate: {}".format(b.p, wr))
    plt.title('Bandit distributions after {} trials'.format(trial))
    plt.legend()
    plt.show()
```

### Experiment Function

```python
def experiment():
    bandits = [Bandit(p) for p in BANDIT_PROBABILITIES]
    
    sample_points = [5, 10, 25, 50, 100, 200, 500, 1000, 1500, 1999]
    rewards = np.zeros(NUM_TRIALS)
    for i in range(NUM_TRIALS):
        # Thompson sampling
        j = np.argmax([b.sample() for b in bandits])
        
        # Plot posteriers at sample points
        if i in sample_points:
            plot(bandits, i)
        
        # Pull for the bandit with largest sample
        x = bandits[j].pull()
        
        # Update reqards
        rewards[i] = x
        
        # Update the distribution for the bandit whose arm was just pulled
        bandits[j].update(x)
        
    # Print total reward
    print("Total reward earned: {}".format(rewards.sum()))
```

```python
experiment()
```
